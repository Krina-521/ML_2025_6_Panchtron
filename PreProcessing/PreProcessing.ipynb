<h4>Dataset imputation with fillna() & interpolation(polynomial) methods.</h4>
 

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# Importing the dataset
df = pd.read_csv('D:/AU/Sem2/ML/Dataset/6.csv')
df.head(3)
```

```python
# Getting the info of the dataset
df.info()
```

```python
# Getting the description of the dataset(Mainly focusing on the Mean & Standard Deviation)
df.describe()
```

```python
# Imputing the missing value using interploate method(poylnomial of order 2)
# Meaning 

imputed_df=df.copy()

imputed_df.interpolate(method='polynomial',order=2, inplace=True)
```

```python
# Imputed dataframe
imputed_df.tail(3)
# imputed_df.isnull().sum()
```

```python
imputed_df.isnull().sum()
```

```python
# For the rows who wasn't get imputed for the missing values
imputed_df.fillna(method='bfill', inplace=True)
```

```python
imputed_df.fillna(method='ffill', inplace=True)
```

```python
imputed_df.head(3)
```

```python
imputed_df.isnull().sum()
```

```python
imputed_df.describe()
```

```python
mean_diff=df.describe().loc['mean']-imputed_df.describe().loc['mean']
std_diff=df.describe().loc['std']-imputed_df.describe().loc['std']

diff_df=pd.DataFrame([mean_diff,std_diff],index=['mean_diff','std_diff'])

print(diff_df.to_string())
```

```python
diff_df_T = diff_df.T
bar_width = 0.4
x = np.arange(len(diff_df_T))

fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # 1 row, 2 columns

# Mean Difference Plot
axes[0].bar(diff_df.columns, diff_df.loc['mean_diff'], color='skyblue', edgecolor='black')
axes[0].axhline(0, color='red', linestyle='--', linewidth=1)   # Add a horizontal line at 0
axes[0].set_title("Mean Differences After Imputation")
axes[0].set_xlabel("Columns")
axes[0].set_ylabel("Mean Difference")
axes[0].tick_params(axis='x', rotation=90)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Standard Deviation Difference Plot
axes[1].bar(diff_df.columns, diff_df.loc['std_diff'], color='lightcoral', edgecolor='black')
axes[1].axhline(0, color='red', linestyle='--', linewidth=1)
axes[1].set_title("Standard Deviation Differences After Imputation")
axes[1].set_xlabel("Columns")
axes[1].set_ylabel("Standard Deviation Difference")
axes[1].tick_params(axis='x', rotation=90)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```

<h4>Dataset imputation with MICE(Multivariate Imputation by Chained Equations).</h4>
 

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputed2_df = df.copy()
imputed2_df.isna().sum()
```

```python
imputed2_df.head(3)
```

```python
# missing_mask=imputed2_df.isna()
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns

# Step 2: Apply IterativeImputer only on numeric columns
imputer = IterativeImputer(max_iter=10, random_state=0)
imputed_values = imputer.fit_transform(df[numeric_cols])

# Step 3: Replace only the numeric missing values in the original DataFrame
df[numeric_cols] = imputed_values

# Step 4: Check results
df.head()

df.to_csv('D:/AU/Sem2/ML/HandsOn Prac/Project/ML_2025_6_Panchtron/Dataset/imputed_MICE.csv',index=False)
```

```python
df.isna().sum()
```

```python

original_df = pd.read_csv('D:/AU/Sem2/ML/HandsOn Prac/Project/ML_2025_6_Panchtron/Dataset/6.csv')
imputed_df1 = pd.read_csv('D:/AU/Sem2/ML/HandsOn Prac/Project/ML_2025_6_Panchtron/Dataset/imputed_interpolate.csv') 
imputed_df2 = pd.read_csv('D:/AU/Sem2/ML/HandsOn Prac/Project/ML_2025_6_Panchtron/Dataset/imputed_MICE.csv')

missing_before = original_df.isna().sum()
missing_after_1 = imputed_df1.isna().sum()
missing_after_2 = imputed_df2.isna().sum()

missing_comparison = pd.DataFrame({
    'Original': missing_before,
    'Imputed1': missing_after_1,
    'Imputed2': missing_after_2
})

# Step 3: Focus only on columns that had missing values initially
missing_comparison = missing_comparison.loc[missing_before[missing_before > 0].index]

# Step 4: Melt for seaborn
missing_comparison = missing_comparison.reset_index().melt(id_vars='index',
                                                          var_name='Stage',
                                                          value_name='Missing Values')

# Step 5: Plot using seaborn
plt.figure(figsize=(14, 6))
sns.barplot(data=missing_comparison, x='index', y='Missing Values', hue='Stage', palette=['#FF6347', '#4682B4', '#32CD32'])

# Step 6: Adjust aesthetics
plt.title('Missing Values: Original vs Imputed1 vs Imputed2', fontsize=16)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Number of Missing Values', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Stage')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### Distribution Of Data

```python
df=pd.read_csv('D:/AU/Sem2/ML/HandsOn Prac/Project/ML_2025_6_Panchtron/Dataset/6.csv')
imputed_interpolate=pd.read_csv('D:/AU/Sem2/ML/HandsOn Prac/Project/ML_2025_6_Panchtron/Dataset/imputed_interpolate.csv')
imputed_MICE=pd.read_csv('D:/AU/Sem2/ML/HandsOn Prac/Project/ML_2025_6_Panchtron/Dataset/imputed_MICE.csv')
```

```python
numeric_cols = df.select_dtypes(include='number').columns

# Plotting
fig, axes = plt.subplots(nrows=len(numeric_cols), ncols=3, figsize=(15, 5 * len(numeric_cols)))
fig.subplots_adjust(hspace=0.3)

datasets = {
    'Original': df,
    'Interpolate Method': imputed_interpolate,
    'MICE': imputed_MICE
}

# Plot each column across all datasets
for i, col in enumerate(numeric_cols):
    for j, (name, dataset) in enumerate(datasets.items()):
        sns.histplot(dataset[col], kde=True, ax=axes[i, j])
        axes[i, j].set_title(f'{col} - {name}')
        axes[i, j].set_xlabel('')

plt.tight_layout()
plt.show()
```
