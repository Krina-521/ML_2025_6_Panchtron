```python
import pandas as pd
import matplotlib.pyplot as plt

columns = athlete_dataset.columns
mean_athlete_dataset = athlete_dataset.mean() # Calculate mean for each column
median_athlete_dataset = athlete_dataset.median() # Calculate median for each column
mode_athlete_dataset = athlete_dataset.mode().iloc[0] # Calculate mode for each column

fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# Scatter Plots
axes[0].scatter(columns, mean_athlete_dataset, color='skyblue', edgecolors='black', s=100)
axes[0].set_title("Mean of Athlete Dataset")

axes[1].scatter(columns, median_athlete_dataset, color='orange', edgecolors='black', s=100)
axes[1].set_title("Median of Athlete Dataset")

axes[2].scatter(columns, mode_athlete_dataset, color='green', edgecolors='black', s=100)
axes[2].set_title("Mode of Athlete Dataset")

# Formatting
for ax in axes:
    ax.set_xlabel("Attributes")
    ax.set_ylabel("Values")
    ax.set_xticks(range(len(columns)))  # Set tick positions
    ax.set_xticklabels(columns, rotation=45)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Dataframe of mean_athlete_dataset
mean_athlete_dataset = pd.DataFrame(mean_athlete_dataset)

plt.figure(figsize=(10, 6))

for col in mean_athlete_dataset.columns:
    sns.kdeplot(mean_athlete_dataset[col], label=col, fill=True, alpha=0.3)

plt.legend()
plt.title("Density Plot of Multiple Columns")
plt.show()
```

```python
# Dataframe of mean_athlete_dataset
median_athlete_dataset = pd.DataFrame(median_athlete_dataset)
plt.figure(figsize=(10, 6))

for col in median_athlete_dataset.columns:
    sns.kdeplot(median_athlete_dataset[col], label=col, fill=True, alpha=0.3)

plt.legend()
plt.title("Density Plot of Multiple Columns")
plt.show()
```
