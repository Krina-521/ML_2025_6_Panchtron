```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer  # MICE imputation
from fancyimpute import IterativeSVD       # EM imputation (via SVD)
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame and 'RSI' is the target variable
target_column = 'RSI'

# 1. Separate the target variable
target = df[target_column]
features = df.drop(columns=[target_column])

# 2. Select numeric columns (exclude the first three non-numeric columns and the target)
numeric_columns = features.columns[3:]
df_numeric = features[numeric_columns]

# Compute observed statistics (using only non-missing values)
observed_mean = df_numeric.mean(skipna=True)
observed_std = df_numeric.std(skipna=True)

# -----------------------------------
# 3. Scale the numeric data for imputation consistency
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_numeric), columns=numeric_columns)

# -----------------------------------
# 4. MICE Imputation using IterativeImputer
mice_imputer = IterativeImputer(max_iter=10, random_state=42)
mice_imputed_scaled = mice_imputer.fit_transform(df_scaled)
mice_imputed = pd.DataFrame(scaler.inverse_transform(mice_imputed_scaled), columns=numeric_columns)

# -----------------------------------
# 5. EM Imputation using IterativeSVD
em_imputer = IterativeSVD(rank=2)
em_imputed_scaled = em_imputer.fit_transform(df_scaled)
em_imputed = pd.DataFrame(scaler.inverse_transform(em_imputed_scaled), columns=numeric_columns)

# -----------------------------------
# 6. Compute imputed statistics for both methods
mice_mean = mice_imputed.mean()
mice_std = mice_imputed.std()
em_mean = em_imputed.mean()
em_std = em_imputed.std()

# Define a function to compute the error percentage
def error_perc(observed, imputed):
    return np.abs((observed - imputed) / observed) * 100

mice_mean_error = error_perc(observed_mean, mice_mean)
mice_std_error = error_perc(observed_std, mice_std)
em_mean_error = error_perc(observed_mean, em_mean)
em_std_error = error_perc(observed_std, em_std)
```

```python

# -----------------------------------
# 7. Create a comparison table
comparison = pd.DataFrame({
    "Feature": numeric_columns,
    "Observed Mean": observed_mean.values,
    "MICE Mean": mice_mean.values,
    "EM Mean": em_mean.values,
    "Observed SD": observed_std.values,
    "MICE SD": mice_std.values,
    "EM SD": em_std.values,
    "MICE Error % (Mean)": mice_mean_error.values,
    "EM Error % (Mean)": em_mean_error.values,
    "MICE Error % (SD)": mice_std_error.values,
    "EM Error % (SD)": em_std_error.values
})

# Function to highlight the minimum error in each row
def highlight_min_error(row):
    styles = [''] * len(row)
    if row["MICE Error % (Mean)"] < row["EM Error % (Mean)"]:
        styles[comparison.columns.get_loc("MICE Error % (Mean)")] = 'font-weight: bold'
    else:
        styles[comparison.columns.get_loc("EM Error % (Mean)")] = 'font-weight: bold'

    if row["MICE Error % (SD)"] < row["EM Error % (SD)"]:
        styles[comparison.columns.get_loc("MICE Error % (SD)")] = 'font-weight: bold'
    else:
        styles[comparison.columns.get_loc("EM Error % (SD)")] = 'font-weight: bold'
    return styles

# Apply the styling to the DataFrame
styled_comparison = comparison.style.apply(highlight_min_error, axis=1)

# Display the styled DataFrame
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
styled_comparison
```
